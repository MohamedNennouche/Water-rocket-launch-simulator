# Example of how to use WaterRocket

```python
from WaterRocket import WaterRocket
```

### 1. Initialization

Add all informations about the rocket and the experimental environment like this

```python
myRocket = WaterRocket(
    bottle_volume= 2, 
    d_bottle= 8.9, 
    d_output= 0.9, 
    m_empty_rocket= 0.5, 
    Cx= 0.1, 
    tilt_angle= 89, 
    length_rampe= 22, 
    initial_pressure= 10, 
    initial_water_volume= 0.65,
    g=9.81,
    r=998,
    ra=1.2,
    Patm=101325
)
```

### 2. Simulate the flight

```python
data_flight = myRocket.create_df(save_as_CSV=False)
data_flight
```

### 3. Main informations about the flight

```python
myRocket.show_flight_infos(save_in_text=False)
```

### 4. Generate flights graphics

```python
myRocket.graphic_all(save_fig=False, show_figure=True)
```

### 5. Generate PDF report about the flight

```python
myRocket.createPDF(path_to_save_pdf="example_report.pdf",saveImgs=False,author="John Doe")
```
